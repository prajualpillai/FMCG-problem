```python
fin = pd.merge(cou,rev2,on="product_id")
fin.head()
```

```python
fin.shape
```

```python
fin.drop(["customer_id","order_id"],axis=1,inplace=True)
fin.head()
```

```python
fin1 = fin[:len(fin)//2-1]
fin2 = fin[len(fin)//2:]
```

```python
def piot(fin):
    piv = fin.pivot_table(index='product_id',columns='review_id',values='review_score').fillna(0)
    return piv
```

```python
mat1 = piot(fin1)
mat2 = piot(fin2)
```

```python
pd.__version__
```

```python
mat3 = mat1.append(mat2)
```

```python
mat3 
```

```python
mat4 = mat3.fillna(0)
```

```python
mat4
```

```python
from scipy.sparse import csr_matrix
```

```python
mat4_sp = csr_matrix(mat4.values)
```

```python
from sklearn.neighbors import NearestNeighbors as nn
```

```python
model = nn(metric = 'cosine',algorithm = 'brute')
model.fit(mat4_sp)
```

```python
qi = np.random.choice(mat4.shape[0])
d,i = model.kneighbors(mat4.iloc[qi,:].values.reshape(1,-1),n_neighbors=6)
```

```python
for j in range(0,len(d.flatten())):
    if j == 0:
        print("Recommendation for: ",mat4.index[qi])
    else:
        print(j,": ",mat4.index[i.flatten()[j]],", distance: ",d.flatten()[j])
```
